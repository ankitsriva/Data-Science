# 1.1) State the Null and Alternate Hypothesis for conducting one-way ANOVA for both the variables ‘A’ and ‘B’ individually. [both statement and statistical form like Ho=mu, Ha>mu]

# Formulate the hypothesis of the One-way ANOVA of 'A' variable

We want to test for the three levels of treatment to see if there’s a difference between the treatments.

$H_0$ : The means with respect to level of treatments for ingredient A is equal.
The null hypothesis states that there is no significant difference between the levels of treatment done for ingredient A 

Ho=mu: mu1=mu2=m3


$H_1$ : At least one of the means with respect to level of treatment for ingredient A is unequal.

Ha=mu: Atleast one of the means is different

The Alternate hypothesis states that there is significant difference for atleast one of the levels of treatment done for ingredient A



# Formulate the hypothesis of the One-way ANOVA of 'B' variable

# Normalizing and Scaling

**We see that the column variables of the data set are of different scales**

**We will use StandardScaler in this method, we convert variables with different scales of measurements into a single scale.**


**StandardScaler normalizes the data using the formula (x-mean)/standard deviation.**

StandardScaler standardizes a feature by subtracting the mean and then scaling to unit variance. Unit variance means dividing all the values by the standard deviation.
</br>

StandardScaler results in a distribution with a standard deviation equal to 1. The variance is equal to 1 also, because variance = standard deviation squared. And 1 squared = 1.
</br>

StandardScaler makes the mean of the distribution 0. About 68% of the values will lie be between -1 and 1.Also for PCA analysis we want to shift the mean towards 0 and do the rotation to capture the Variances in data.

# 2.8) Mention the business implication of using the Principal Component Analysis for this case study.

Principal component analysis helps extracting the features, we can drop the least important variables while still retaining the most valuable parts of all of the variables capturing most of the variances.

Inferences from the above represent the directions in which the data has maximum variance and also the directions in which the data is most spread out.

This heatmap and the color bar basically represent the correlation between the various feature and the principal component itself

Cumulative % gives the percentage of variance accounted for by the n components. For example, the cumulative percentage for the second component is the sum of the percentage of variance for the first and second components. It helps in deciding the number of components by selecting the components which explained the high variance.

In the above observation we see that the first feature explains 32% of the variance within our data set while the first two explain 58.3 and so on. If we employ 9 features we capture ~ 90%+ of the variance within the dataset, thus we gain very little by implementing an additional feature (think of this as diminishing marginal return on total variance explained).

For the Education dataset we see that we have total 17 continuous variable and doing an analysis is tough, hence we reduced the dimension from 17 to 9 so that we will have the 90% of data for analyses and with little data loss. It helped us to understand the students enrolling into the colleges for education with less number of characteristics that is PC containing most of the data.
